```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,129]
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins=3)
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,160], rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,160], rwidth=0.95, color='g')
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel('Sugar Range')
plt.ylabel('total number of patients')
plt.title('Blood sugar Analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150] ,label=['men', 'women'])
plt.legend()
```
