```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ['Home Rent', 'Food', 'Phone/Internet Bill', 'Car','Other Utilities']
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=2)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0.1,0.1])
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0.1,0.1], startangle=45)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0.1,0.1], startangle=180)
plt.show()
```
