Matplotlib is a python 2D plotting lib which produces quality figures in a variety if hardcopy formats

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,43,44,49]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Graph')
plt.plot(x,y, color='red',linewidth=5, linestyle='dotted',)
```

format strings in plot functions

```python
plt.plot(x,y, 'rD--')
```

```python
plt.plot(x,y, color='red', marker='D')
```

```python
plt.plot(x,y, color='blue', alpha=0.6) ## aplha is for transparency
```
