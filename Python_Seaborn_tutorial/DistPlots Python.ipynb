DisPlots -> Distribution Plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Loading the Penguins Datasets

```python
pen = sns.load_dataset('penguins')
```

```python
pen.head()
```

Drawing a Simple distplot

```python
sns.distplot(pen.flipper_length_mm) 
```

```python
sns.distplot(pen.flipper_length_mm, rug = True)
```

Removing the KDE Plot

```python
sns.distplot(pen.flipper_length_mm, kde = False)
```

```python
sns.distplot(pen.flipper_length_mm, hist = False)
```

```python
sns.distplot(pen.flipper_length_mm, rug_kws = {'color':'red', 'height':0.30} ,rug = True)
```

```python
sns.distplot(pen.flipper_length_mm, kde = False ,hist_kws = {'color':'red','histtype':'stepfilled'})
```

```python
sns.distplot(pen.flipper_length_mm, hist = False ,kde = True ,kde_kws = {'color':'red','lw':4.0,'alpha':0.9})
```
